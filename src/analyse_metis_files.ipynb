```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_mine = "metis_files/graph_metis_weights_george.txt"
path_simu = "metis_files/graph_metis_events_as_weights.txt"
```

```python
def parse_metis_data(path):
    weights = []
    with open(path, 'r') as f:
        header = f.readline() # skip header
        print(f'skipping header: {header}')
        for line in f:
            weights += [int(line.split(' ')[0])]
    return np.array(weights)

def normalise(values):
    return (values - np.min(values)) / (np.max(values) - np.min(values))
```

```python
mine = parse_metis_data(path_mine)
simu = parse_metis_data(path_simu)

mine = normalise(mine)
simu = normalise(simu)
```

```python
plt.figure(figsize=(6,3))

plt.hist(mine, bins=100, histtype='step', linewidth=1.5, label='python analysis')
plt.hist(simu, bins=100, histtype='step', linewidth=1, label='simulator')

plt.yscale('log')
plt.grid(alpha=.4)
plt.xlabel('normalised weights')
plt.ylabel('Num. Switches')
plt.title('Python VS Simulator: Weights for partition')
plt.legend()
plt.tight_layout()
plt.show()
```
